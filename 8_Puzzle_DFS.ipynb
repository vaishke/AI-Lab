# 8-Puzzle Problem

Given a 3x3 grid with 8 numbered tiles (1 - 8) and 1 blank space (_).

There is a final goal state and an initial state of the board given. The problem is to search the solution such that the tiles in the initial state of the board are moved such that they reach the given goal state.

Total no of board possibilities = 9!

Some of the states are unsolvable i.e. no matter what the goal state cannot be reached. So we need to ignore/remove such states.


```python
def is_goal(goal: list[list[str]], board: list[list[str]]) -> bool:
    return goal == board

def find_empty_tile(board: list[list[int]]) -> list[int]:
    ans = list()
    for i in range(3):
        for j in range(3):
            if board[i][j] == "_":
                ans = [i, j]
                break
    return ans

def swap_tiles(board: list[list[int]], x: int, y: int, x1: int, y1: int) -> list[list[int]]:
    new_board = [row[:] for row in board]
    new_board[x][y], new_board[x1][y1] = new_board[x1][y1], new_board[x][y]
    return new_board

def dfs(goal: list[list[int]], board: list[list[int]], visited: set[tuple]) -> bool:
    print(board)
    
    if is_goal(goal, board):
        print("GOAL REACHED")
        return True
    
    board_tuple = tuple(tuple(row) for row in board)
    if board_tuple in visited:
        return False
    
    visited.add(board_tuple)   
    [x, y] = find_empty_tile(board)
    
    dx = [0, -1, 0, 1]
    dy = [-1, 0, 1, 0]

    for i in range(4):
        new_x = x + dx[i]
        new_y = y + dy[i]

        if 0 <= new_x < 3 and 0 <= new_y < 3:
            new_state = swap_tiles(board, x, y, new_x, new_y)
            if dfs(goal, new_state, visited):
                return True
    return False

    

def main():
    print("8-puzzle problem")

    goal_state = [["_"] * 3 for _ in range(3)]
    initial_state = [["_"] * 3 for _ in range(3)]

    print("Enter goal state: ")
    for i in range(3):
        goal_state[i] = list(input().split())
    print("Enter initial state: ")
    for i in range(3):
        initial_state[i] = list(input().split())
    
    visited = set()
    dfs(goal_state, initial_state, visited)
    

main()
```
