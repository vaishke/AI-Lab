```python
data = [[1, 3], [2, 5], [3, 7], [4, 9], [5, 11], [6, 13], [7, 14], [8, 17], [9, 18],
    [10, 21], [11, 23], [12, 24], [13, 27], [14, 28], [15, 31]]
```

```python
# y = ax + b

N = len(data)
SUM_X = sum(d[0] for d in data)
SUM_X2 = sum(d[0] * d[0] for d in data)
SUM_Y = sum(d[1] for d in data)
SUM_Y2 = sum(d[1] * d[1] for d in data)
SUM_XY = sum(d[0] * d[1] for d in data)

def find_slope() -> float:
    return ((N * SUM_XY) - (SUM_X * SUM_Y)) / ((N * SUM_X2) - SUM_X ** 2)

def find_y_intercept() -> float:
    return ((SUM_Y * SUM_X2) - (SUM_X * SUM_XY)) / ((N * SUM_X2) - SUM_X ** 2)

def find_y(x: int) -> float:
    return x * find_slope() + find_y_intercept()

print(f"Slope = {find_slope()}\n")
print(f"Y-Intercept = {find_y_intercept()}\n")
print(f"If x = 20, then y = {find_y(20)}\n")
```
